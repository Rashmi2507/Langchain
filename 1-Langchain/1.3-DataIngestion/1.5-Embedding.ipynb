Converting text into vector

```python
import os
from dotenv import load_dotenv
load_dotenv() # load all the env variables from the .env file
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is a sample text to be embedded."
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimension=1024)
```

```python
embeddings_1024
```

```python
text = "This is a sample text to be embedded."
query_result_1024 = embeddings_1024.embed_query(text)
query_result_1024
```

```python
len(query_result_1024)
```

```python
## Text Loader

from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
docs = loader.load()
docs
```

```python
%pip install langchain-text-splitters -q
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=20)
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
## Combine Vector Embedding and vector StoreDB

from langchain_community.vectorstores import Chroma

# convert the documents into text
db = Chroma.from_documents(final_docs, embeddings_1024) 
db #VectorStoreDB object
```

```python
## Retrieved the result from querying the vector storeDB using similarity search

query = "First, let's talk about Models. Langchain supports multiple LLM providers"
retrieved_result = db.similarity_search(query)
print(retrieved_result)
```
