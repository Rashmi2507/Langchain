### HuggingFace Embeddings
Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text and image embeddings. One of the embedding models is used in the Hugging face Embeddings class. We have also added an alias for sentence Transformer Embeddings for users who are more familiar with directly using that package

```python
from langchain_huggingface import HuggingFaceEmbeddings

# Sentence Transformers model for embeddings
HuggingFaceEmbeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
text = "This is a test sentence for generating embeddings using Hugging Face's Sentence Transformers model."
query_result = HuggingFaceEmbeddings.embed_query(text)
print(query_result)
```

```python
len(query_result)
```

```python
doc_result = HuggingFaceEmbeddings.embed_documents([text, "this is another test sentence for generating embeddings using Hugging Face's Sentence Transformers model."])
print(doc_result[0])
```
